# Why Choose Matplotlib and Seaborn?

## Basic of Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y)
```

```python
categories = ['A', 'B', 'C']
values = [10, 20, 15]
colors = ['skyblue', 'lightcoral', 'lightgreen']
plt.bar(categories, values, color=colors)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# Bar Chart
categories = ['A', 'B', 'C']
values = [10, 20, 15]
plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

## Visualizations with Seaborn

### Histogram and KDE plot

```python
import seaborn as sns
```

```python
data = np.random.randn(1000)
sns.histplot(data, kde=True, color='Green')
plt.title("Histogram with KDE")
plt.show()
```

### Heatmap

```python
matrix = np.random.rand(5, 5)
sns.heatmap(matrix, annot=True, cmap="prism")
plt.title("Heatmap Example")
plt.show()
```

### Pair Plot

```python
sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species", markers=["o", "s", "D"])
plt.suptitle("Pair Plot Example", y=1.02)
plt.show()
```

### Task

```python
# read ice cream rating.csv
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Pandas/Ice Cream Ratings.csv")
df
```

```python
# set date as index
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
print(df)
```

```python
# scatter plot using texture rating and overall rating
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Texture Rating', y='Overall Rating', data=df, color='royalblue', alpha=0.8)
plt.title('Scatter Plot of Texture Rating vs. Overall Rating (Seaborn)')
plt.xlabel('Texture Rating')
plt.ylabel('Overall Rating')
plt.grid(True)
plt.show()
```

```python
# hist plot
plt.figure(figsize=(8, 6))
sns.histplot(df['Overall Rating'], kde=True, color='lightgreen')
plt.title('Distribution of Overall Ratings (Seaborn)')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# box plot
df["Texture Rating"].plot.box()
```

```python
# pie chart
df["Texture Rating"].value_counts().plot.pie()
```

```python
# plot bar
df["Texture Rating"].value_counts().plot.bar()
```
